## Diabetes Predictions

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
print(dataset['DESCR'])
```

```python
dataset
```

```python
df = pd.DataFrame(dataset.data, columns = ['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
```

```python
df.head()
```

```python
X = df
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot=True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
```

## Hyperparameter Tuning

```python
params = {
    'criterion' : ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3,4,5,10,15,20,25],
    'max_features' : ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
grid = GridSearchCV(dtr, param_grid = params, cv=5, scoring='neg_mean_squared_error')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
import warnings
warnings.filterwarnings('ignore')
model = DecisionTreeRegressor(criterion= 'absolute_error',max_depth=2,max_features= 'log2',splitter= 'best')
model.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model, filled=True)
plt.show()
```

```python
selectedmodel=DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='log2',splitter='random')
selectedmodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectedmodel,filled=True)
plt.show()
```

```python
y_pred1 = model.predict(X_test)
print(r2_score(y_test, y_pred1))
print(mean_squared_error(y_test, y_pred1))
print(mean_absolute_error(y_test, y_pred1))
```

```python
y_pred2 = selectedmodel.predict(X_test)
print(r2_score(y_test, y_pred2))
print(mean_squared_error(y_test, y_pred2))
print(mean_absolute_error(y_test, y_pred2))
```

```python
import warnings
warnings.filterwarnings('ignore')
model2 = DecisionTreeRegressor(criterion= 'poisson',max_depth=5,max_features= 'sqrt',splitter= 'random')
model2.fit(X_train, y_train)
```

```python
y_pred3 = model2.predict(X_test)
print(r2_score(y_test, y_pred3))
print(mean_squared_error(y_test, y_pred3))
print(mean_absolute_error(y_test, y_pred3))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model2,filled=True)
plt.show()
```
