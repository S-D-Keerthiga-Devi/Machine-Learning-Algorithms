```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
print(iris["DESCR"])
```

```python
print(iris['target'])
```

```python
## Independent and Dependent features
X = pd.DataFrame(iris['data'], columns = ['sepal length', 'sepal width', 'petal length', 'petal width'])
y = iris['target']
```

```python
 # Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 , random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
treeC = DecisionTreeClassifier()
treeC.fit(X_train, y_train)
```

```python
## Visualization
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treeC, filled = True)
plt.show()
```

```python
treeC1 = DecisionTreeClassifier(max_depth = 2)
treeC1.fit(X_train, y_train)
```

```python
## Visualization
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treeC1, filled = True)
plt.show()
```

```python
y_pred = treeC1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_pred, y_test))
print(classification_report(y_pred, y_test))
```

```python
y_pred1 = treeC.predict(X_test)
print(confusion_matrix(y_pred1, y_test))
print(classification_report(y_pred1, y_test))
```

## Decision Tree Pre pruning and Hyperparameter Tuning

```python
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(treeC, param_grid = params, cv=5, scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
from sklearn.metrics import accuracy_score
y_pred2 = grid.predict(X_test)
print(confusion_matrix(y_pred2, y_test))
print(accuracy_score(y_pred2, y_test))
print(classification_report(y_pred2, y_test))
```
